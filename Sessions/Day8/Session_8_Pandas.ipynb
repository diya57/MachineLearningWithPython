## What is Pandas?

Pandas is a powerful Python library used for data manipulation and analysis. It provides data structures like **Series** and **DataFrame** to handle labeled and relational data efficiently. Pandas is widely used in data science and machine learning workflows.

**Key Features of Pandas:**
- Fast and efficient data manipulation
- Integrated handling of missing data
- Alignment of data for operations
- Data visualization support


```python
# Install and import Pandas
import pandas as pd
print('Pandas version:', pd.__version__)
```

## Introduction to Pandas Series

A **Pandas Series** is a one-dimensional labeled array capable of holding any data type. It is similar to a column in a spreadsheet but more powerful due to its labels (index).

**Key Characteristics:**
- Homogeneous data (all elements of the same type).
- Labeled index (row labels).
- Supports various data types (integers, floats, strings, etc.).

**Creating a Series:**

```python
# Creating a Pandas Series
import pandas as pd

# Creating a Series from a list
data = [10, 20, 30, 40, 50]
series = pd.Series(data)
print(series)
```

### Attributes of a Series

- **index**: Returns the index (row labels).
- **values**: Returns the data values.
- **dtype**: Returns the data type of elements in the Series.


```python
# Exploring Series attributes
print('Index:', series.index)
print('Values:', series.values)
print('Data Type:', series.dtype)
```

### Creating a Series with Custom Indices

You can assign custom labels to the indices when creating a Series. This makes data access easier and more intuitive.

```python
# Creating a Series with custom indices
data = [100, 200, 300]
labels = ['a', 'b', 'c']
custom_series = pd.Series(data, index=labels)
print(custom_series)
```

## Accessing Elements in a Series

You can access elements in a Series using:
- **Index labels**: Use the custom or default index.
- **Integer indices**: Similar to lists in Python.

**Examples:**

```python
# Accessing elements
print('Element with index a:', custom_series['a'])
print('Element at position 1:', custom_series[1])
```

## Basic Operations on Series

Pandas Series supports various operations:
- Element-wise operations (e.g., addition, multiplication).
- Applying NumPy functions.
- Filtering based on conditions.

```python
# Basic operations on Series
series = pd.Series([1, 2, 3, 4, 5])
print('Original Series:\n', series)

# Element-wise operations
print('\nSeries + 10:\n', series + 10)
print('\nSeries * 2:\n', series * 2)

# Applying a NumPy function
import numpy as np
print('\nSquare root:\n', np.sqrt(series))
```

## Series Methods

Pandas Series comes with a variety of built-in methods to make data manipulation easier. Here are some commonly used methods:

- **`head()`**: Displays the first n rows.
- **`tail()`**: Displays the last n rows.
- **`describe()`**: Provides summary statistics.
- **`value_counts()`**: Counts unique values.
- **`sort_values()`**: Sorts the Series.
- **`apply()`**: Applies a function to each element in the Series.

Let's explore these methods with examples.

```python
# Sample Series
data = [50, 20, 30, 40, 10]
labels = ['a', 'b', 'c', 'd', 'e']
series = pd.Series(data, index=labels)

# Using Series methods
print('First 3 elements:\n', series.head(3))
print('\nLast 2 elements:\n', series.tail(2))
print('\nSummary statistics:\n', series.describe())
```

```python
# Counting unique values and sorting
data = ['apple', 'banana', 'apple', 'orange', 'banana', 'apple']
series = pd.Series(data)

print('Value counts:\n', series.value_counts())
print('\nSorted values:\n', series.sort_values())
```

### Applying Functions with `apply()`

The `apply()` method allows you to apply custom or built-in functions to each element in the Series.

```python
# Using apply() to apply a lambda function
series = pd.Series([1, 2, 3, 4, 5])
print('Original Series:\n', series)

# Squaring each element
squared = series.apply(lambda x: x ** 2)
print('\nSquared Series:\n', squared)
```

## Series with Python Functionalities

Pandas Series can be combined with native Python functionalities for tasks such as:
- Iterating through elements.
- Performing mathematical operations.
- Using conditionals.

Let's look at some examples:

```python
# Iterating through a Series
series = pd.Series([10, 20, 30, 40])
for value in series:
    print('Value:', value)
```

```python
# Using conditionals
series = pd.Series([10, 20, 30, 40])

# Elements greater than 20
filtered = series[series > 20]
print('Filtered Series:\n', filtered)
```

## Boolean Indexing on Series

Boolean indexing allows you to filter data based on conditions. It returns a subset of the Series where the condition is `True`.

**Examples of Boolean Indexing:**

```python
# Boolean indexing example
series = pd.Series([10, 15, 20, 25, 30])

# Condition: Values greater than 20
print('Values greater than 20:\n', series[series > 20])
```

### Compound Conditions

You can combine multiple conditions using logical operators such as:
- **`&`**: Logical AND
- **`|`**: Logical OR

For example:

```python
# Compound condition example
series = pd.Series([10, 15, 20, 25, 30])

# Condition: Values greater than 10 and less than 30
filtered = series[(series > 10) & (series < 30)]
print('Filtered values:\n', filtered)
```

## Plotting with Matplotlib

Matplotlib provides a simple interface to create visualizations from Pandas Series. Here are some commonly used plot types:

- **Line plot**: Useful for continuous data.
- **Bar plot**: Ideal for categorical data.
- **Histogram**: Displays the distribution of data.

Let's explore these plots with examples.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample Series
data = [10, 20, 15, 30, 25]
labels = ['A', 'B', 'C', 'D', 'E']
series = pd.Series(data, index=labels)

# Line plot

series.plot(kind='pie', title='Line Plot')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample Series with more meaningful data
data = [10, 20, 15, 30, 25]
labels = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
series = pd.Series(data, index=labels)

# Create a figure with larger size
plt.figure(figsize=(10, 6))

# Create two subplots
plt.subplot(1, 2, 1)
# Line plot
series.plot(kind='line', marker='o', linewidth=2, markersize=8)
plt.title('Sales Trend by Product', fontsize=12, pad=15)
plt.xlabel('Products', fontsize=10)
plt.ylabel('Sales Units', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)

plt.subplot(1, 2, 2)
# Pie chart
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#ff99cc']
series.plot(kind='pie', autopct='%1.1f%%', colors=colors)
plt.title('Sales Distribution', fontsize=12, pad=15)

# Adjust layout to prevent overlap
plt.tight_layout()

# Add a super title
plt.suptitle('Product Sales Analysis', fontsize=14, y=1.05)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample Series with more meaningful data
data = [100, 20, 15, 30, 25]
labels = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']

# Calculate percentages
total = sum(data)
percentages = [round((x/total)*100, 1) for x in data]

# Create Series
series = pd.Series(percentages, index=labels)

# Create figure with a good size
plt.figure(figsize=(10, 6))

# Create bar chart
bars = series.plot(kind='bar', 
                  color='skyblue',
                  edgecolor='navy')

# Add value labels on top of each bar
for i, v in enumerate(percentages):
    plt.text(i, v + 0.5, f'{v}%', 
             ha='center', 
             fontweight='bold')

# Customize the chart
plt.title('Product Distribution by Percentage', 
         fontsize=14, 
         pad=15)
plt.xlabel('Products', fontsize=12)
plt.ylabel('Percentage (%)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Adjust layout
plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate survival statistics by gender
survival_by_gender = pd.crosstab(titanic_df['sex'], titanic_df['survived'])
survival_by_gender.columns = ['Dead', 'Survived']

# Set the style
plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(12, 8))

# Create the bar chart
ax = survival_by_gender.plot(kind='bar',
                           width=0.8,
                           color=['#FF6B6B', '#4ECDC4'],  # Coral red for dead, Turquoise for survived
                           edgecolor='black',
                           linewidth=1.5)

# Customize the chart
plt.title('Survival Distribution by Gender on Titanic', 
         fontsize=16, 
         pad=20, 
         fontweight='bold',
         color='#2C3E50')  # Dark blue title

# Add labels
plt.xlabel('Gender', fontsize=12, fontweight='bold', color='#2C3E50')
plt.ylabel('Number of Passengers', fontsize=12, fontweight='bold', color='#2C3E50')

# Customize grid
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add value labels on the bars
for i in ax.containers:
    ax.bar_label(i, 
                 label_type='center',
                 fontsize=11,
                 fontweight='bold',
                 color='white')
    ax.bar_label(i, 
                 label_type='edge',
                 fontsize=11,
                 fontweight='bold',
                 padding=3)

# Calculate and display percentages
total_males = survival_by_gender.loc['male'].sum()
total_females = survival_by_gender.loc['female'].sum()

for i, gender in enumerate(['male', 'female']):
    total = survival_by_gender.loc[gender].sum()
    dead_pct = (survival_by_gender.loc[gender, 'Dead'] / total) * 100
    survived_pct = (survival_by_gender.loc[gender, 'Survived'] / total) * 100
    
    # Add percentage annotations
    plt.text(i, survival_by_gender.loc[gender, 'Dead']/2, 
             f'{dead_pct:.1f}%\nDead', 
             ha='center', 
             color='white',
             fontweight='bold')
    
    plt.text(i, survival_by_gender.loc[gender, 'Dead'] + survival_by_gender.loc[gender, 'Survived']/2,
             f'{survived_pct:.1f}%\nSurvived', 
             ha='center', 
             color='white',
             fontweight='bold')

# Add a summary box
summary_text = f"""Total Passengers: {len(titanic_df):,}
Male Passengers: {total_males:,}
- Dead: {survival_by_gender.loc['male', 'Dead']:,} ({survival_by_gender.loc['male', 'Dead']/total_males*100:.1f}%)
- Survived: {survival_by_gender.loc['male', 'Survived']:,} ({survival_by_gender.loc['male', 'Survived']/total_males*100:.1f}%)

Female Passengers: {total_females:,}
- Dead: {survival_by_gender.loc['female', 'Dead']:,} ({survival_by_gender.loc['female', 'Dead']/total_females*100:.1f}%)
- Survived: {survival_by_gender.loc['female', 'Survived']:,} ({survival_by_gender.loc['female', 'Survived']/total_females*100:.1f}%)"""

plt.text(1.45, plt.ylim()[1]*0.95, summary_text,
         bbox=dict(facecolor='white', 
                  alpha=0.9,
                  edgecolor='gray',
                  boxstyle='round,pad=1'),
         fontsize=10,
         verticalalignment='top')

# Customize legend
plt.legend(title='Status',
          title_fontsize=12,
          fontsize=10,
          bbox_to_anchor=(1.15, 0.6),
          edgecolor='black')

# Rotate x-axis labels
plt.xticks(rotation=0)

# Adjust layout
plt.tight_layout()

plt.show()
```
